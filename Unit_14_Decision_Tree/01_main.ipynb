```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# Tennis dataset
dataset = pd.DataFrame({
    'Outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild','Cool'],
    'Humidity': ['High','High','High','High','Normal','Normal','High','High','Normal','Normal','Normal','High','Normal','High'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong'],
    'Play': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
})

# Encode categorical variables
# Outlook:      Overcast=0,     Rain=1,     Sunny=2
# Temperature:  Cool=2,         Hot=1,      Mild=0
# Humidity:     High=0,         Normal=1
# Wind:         Strong=1,       Weak=0
# PlayTennis:   No=0,           Yes=1

le = LabelEncoder()
for column in dataset.columns:
    dataset[column] = le.fit_transform(dataset[column])
dataset
```

```python
# Extract input and output arrays
X = dataset[['Outlook', 'Temperature', 'Humidity', 'Wind']]
Y = dataset['Play']

# Train the model
model = DecisionTreeClassifier()
model.fit(X, Y)

# Test the model
Score = model.score(X, Y) * 100
print(f"{Score = :5.2f}%")

# Plot the tree
plt.figure(figsize=(8,12))
plot_tree(model, feature_names=X.columns, class_names=['No','Yes'])
plt.show()
```
