```python
from sklearn.datasets import load_iris
from KNN_module import Point
from KNN_module import KNN
import pandas as pd

# Load the dataset
iris = load_iris()
print(f"Dataset features: {iris.feature_names}") # Features names
print(f"Target Species  : {iris.target_names}")  # Species names

# Convert the dataset into a dataframe
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a list of points
points = []
for i in range(len(iris_df)):
    point = Point(
        vals = [iris_df.iloc[i]['sepal length (cm)'], iris_df.iloc[i]['sepal width (cm)']],
        label = iris_df.iloc[i]['species']
    )
    points.append(point)

# Create a KNN object
knn = KNN(k=3)
knn.update_points(points)

x = 6.5     #float(input("Enter sepal length (cm): "))
y = 3.3     #float(input("Enter sepal width (cm) : "))
predicted_class = knn.predict(Point([x, y], None))
print(f"{predicted_class = }")
```
